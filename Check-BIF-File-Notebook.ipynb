```python
class RecordItem:
    
    def __init__(self,id):
        self.id = id 
        self.listparam = []

    def RecordId(self):
        return self.id
    def AddParameter(self,parameter):
        self.listparam.append(parameter)
        return
    def PrintRecordContents (self):
        print (self.id,self.listparam)
    def GetParam (self):
        return self.listparam
```

```python
class GroupItem:
    
    def __init__(self,name):
        self.name = name
        self.listrecord = []
    def GroupName(self):
        return self.name
    def AddRecord(self,record):
        self.listrecord.append(record)
        return
    def PrintGroupContents (self):
        print (self.name,self.listrecord)
    def CheckRecord (self, recordid):
        for i in self.listrecord:
            if (i.RecordId() == recordid): 
               # print ("already in record list")
                return True
        #print("not already in record list")
        return False
    def GetRecord (self,record):
        for i in self.listrecord:
            if (i.RecordId() == record): 
                #print ("already in record list")
                return i
    
        #print("not already in record list")
        return None
    def GetAllRecords (self):
        return self.listrecord
```

```python
class SiteItem:
    
    def __init__(self,id):
        self.id = id
        self.listgroup = []
    def SiteId(self):
        return self.id
    def AddGroup(self,group):
        self.listgroup.append(group)
        return
    def PrintSiteContents (self):
        print (self.id ," : ")
        for i in self.listgroup : print (i.GroupName(),",",end='')
    def PrintSiteId (self):
        print(self.id)
    def CheckGroup (self,group):
        for i in self.listgroup:
            if (i.GroupName() == group): 
               # print ("already in group list")
                return True
    
        #print("not already in group list")
        return False
    def GetGroup (self,group):
        for i in self.listgroup:
            if (i.GroupName() == group): 
               # print ("already in group list")
                return i
    
        #print("not already in group list")
        return None
    def CheckGroupOLD (self,group):
        if group in self.listgroup:
            #print ("already in group list")
            return True
        else:
           # print("not already in group list")
            return False
```
